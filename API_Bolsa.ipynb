```python
client = Brapi(api_key='YOUR_TOKEN')#, params= 'defaultKeyStatistics,defaultKeyStatisticsHistoryQuarterly')

quotes = client.quote.retrieve(
    tickers='PETR4', #KIVO11, KNSC11, SPTW11, VGRI11, GCRA11
    range='5d',
    interval='1d',
    dividends=True,
    modules='defaultKeyStatistics,defaultKeyStatisticsHistoryQuarterly',
    fundamental=True
)

print(quotes.results[0].regular_market_price)
quotes
```

```python
client = Brapi(api_key="seu_token")

def get_quotes_dataframe(tickers: list[str]) -> pd.DataFrame:
    """Retorna cotações como DataFrame"""
    tickers_str = ','.join(tickers)
    quote = client.quote.retrieve(tickers=tickers_str)
    
    data = []
    for stock in quote.results:
        data.append({
            'symbol': stock.symbol,
            'name': stock.short_name,
            'price': stock.regular_market_price,
            'change_percent': stock.regular_market_change_percent,
        })
    
    return pd.DataFrame(data)

# Uso
df = get_quotes_dataframe(['PETR4', 'VALE3', 'ITUB4'])
print(df)

# Salvar em Excel
# df.to_excel('cotacoes.xlsx', index=False)
```
